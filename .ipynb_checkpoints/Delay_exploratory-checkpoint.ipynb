```python
def clean_df(df):
    
    # create new datetime object
    df['dt_local_clean'] = df['dt_local_clean'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S"), 1)

    # generate day of week dummies
    dow = pd.get_dummies(df['dow'])
    
    # generate date dummies
    date = pd.get_dummies(df['date'], prefix='date')
    
    # generate odid dummies
    odid = pd.get_dummies(df['odid'], prefix='odid')
    
    # concat
    df = pd.concat([df, dow, date, odid],axis=1)
    df.rename(columns={1:'Monday', 2:'Tuesday', 3:'Wednesday', 4:'Thursday', 5:'Friday'}, inplace=True)
    
    return df
```

```python
# clean the df's
delay_bang = clean_df(delay_bang)
delay_delhi = clean_df(delay_delhi)
```

```python
# unconcat bangalore df's
delay_bang_train = delay_bang[delay_bang['odid']<=177]
delay_bang_test = delay_bang[delay_bang['odid']>=178]
```

```python
best_alpha = 0.0
best_score = 0.0
best_mlp = MLPRegressor()
```

```python
for alpha in [0.0001, 0.0003, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3]:
    
    print("MLP with alpha=" + str(alpha))
    mlp = MLPRegressor(alpha=alpha)
    mlp_fit = mlp.fit(X_train[date_cols + ['predicted_delay']], y_train)
    print("Fitted MLP")
    mlp_score = mlp.score(X_test[date_cols + ['predicted_delay']], y_test)
    print("Score: " + str(mlp_score))
    
    if mlp_score > best_score:
        best_score = mlp_score
        best_mlp = mlp_fit
        best_alpha = alpha
```

```python
for alpha in [10, 30]:
    
    print("MLP with alpha=" + str(alpha))
    mlp = MLPRegressor(alpha=alpha)
    mlp_fit = mlp.fit(X_train[date_cols + ['predicted_delay']], y_train)
    print("Fitted MLP")
    mlp_score = mlp.score(X_test[date_cols + ['predicted_delay']], y_test)
    print("Score: " + str(mlp_score))
    
    if mlp_score > best_score:
        best_score = mlp_score
        best_mlp = mlp_fit
        best_alpha = alpha
```

```python
print("Best MLP with alpha=" + str(best_alpha))
print("Adj R2 of prediction on test set: " + str(best_score))
```

# Combined Model

1. Predicted delay is a pretty linear predictor of live delay.
2. Date FEs and HFF FEs are not linear predictors of the residuals generated from regression of live on pred delay. (analysis in Stata)
3. Is it possible that they are NONLINEAR predictors of the residuals?
*answer: seems to be no*

```python
import statsmodels.api as sm
```

```python
# Save the residuals from OLS reg of delay on predicted delay - Train set
lr_train = sm.OLS(y_train, X_train['predicted_delay'], hasconst=True).fit()
residuals_train = lr_train.resid
```

```python
# Save the residuals from OLS reg of delay on predicted delay - Test set
lr_test = sm.OLS(y_test, X_test['predicted_delay'], hasconst=True).fit()
residuals_test = lr_test.resid
```

Predict residuals using date FEs with a linear model

```python
# sklearn model: predict residuals using date FEs
lin_predict = LinearRegression()
```

```python
lin_predict.fit(X_train[date_cols], residuals_train.reshape(-1,1))
```

```python
# Adj R2 of the test prediction 
lin_predict.score(X_test[date_cols], residuals_test.reshape(-1,1))
```

Predict residuals using date FEs with a random forest

```python
rf_predict = RandomForestRegressor()
```

```python
rf_predict.fit(X_train[date_cols], residuals_train)
```

```python
# Adj R2 of the test prediction 
rf_predict.score(X_test[date_cols], residuals_test)
```

Predict residuals using date FEs with an MLP

```python
mlp_predict = MLPRegressor(alpha=0.1)
```

```python
mlp_predict.fit(X_train[date_cols], residuals_train)
```

```python
# Adj R2 of the test prediction 
mlp_predict.score(X_test[date_cols], residuals_test)
```
